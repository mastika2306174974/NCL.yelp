#3. Handling Outliers:

```python
import pandas as pd
import numpy as np
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path menuju file CSV yang telah dipreproses dan disimpan di Google Drive
file_path = '/content/drive/My Drive/preprocessed_data PSD.csv'

# Baca data CSV yang telah dipreproses
preprocessed_data = pd.read_csv(file_path)

# Identifikasi dan penanganan outlier untuk kolom numerik
numeric_columns = preprocessed_data.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_columns:
    # Hitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
    Q1 = preprocessed_data[col].quantile(0.25)
    Q3 = preprocessed_data[col].quantile(0.75)

    # Hitung interquartile range (IQR)
    IQR = Q3 - Q1

    # Tentukan batas atas dan batas bawah untuk outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifikasi outlier
    outliers = preprocessed_data[(preprocessed_data[col] < lower_bound) | (preprocessed_data[col] > upper_bound)]

    # Tangani outlier
    # Misalnya, mengganti nilai outlier dengan nilai batas atas atau batas bawah
    preprocessed_data.loc[(preprocessed_data[col] < lower_bound), col] = lower_bound
    preprocessed_data.loc[(preprocessed_data[col] > upper_bound), col] = upper_bound

# Tampilkan data setelah menangani outlier
print("Data setelah menangani outlier:")
print(preprocessed_data.head())

```

#4. Handling Missing Values:

```python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path menuju file CSV yang telah dipreproses dan disimpan di Google Drive
file_path = '/content/drive/My Drive/preprocessed_data PSD.csv'

# Baca data CSV yang telah dipreproses
preprocessed_data = pd.read_csv(file_path)

# Tampilkan jumlah nilai yang hilang (missing values) dalam setiap kolom
missing_values = preprocessed_data.isnull().sum()
print("Jumlah nilai yang hilang dalam setiap kolom:")
print(missing_values)

# Tangani nilai yang hilang
# Misalnya, mengganti nilai yang hilang dengan nilai rata-rata atau median dari kolom tersebut
# Anda juga dapat menggunakan teknik imputasi lainnya sesuai dengan kebutuhan analisis Anda

# Contoh: Mengganti nilai yang hilang dengan nilai median untuk setiap kolom numerik
numeric_columns = preprocessed_data.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_columns:
    median_value = preprocessed_data[col].median()
    preprocessed_data[col].fillna(median_value, inplace=True)

# Contoh: Mengganti nilai yang hilang dengan nilai yang paling sering muncul (mode) untuk setiap kolom kategorikal
categorical_columns = preprocessed_data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    mode_value = preprocessed_data[col].mode()[0]
    preprocessed_data[col].fillna(mode_value, inplace=True)

# Tampilkan data setelah menangani nilai yang hilang
print("\nData setelah menangani nilai yang hilang:")
print(preprocessed_data.head())
```
